1. Напишите запрос, который позволит узнать, какие работодатели находятся на первом и пятом месте по количеству вакансий.

```python
query_5_1 = f'''
-- Определяем работодателей, занимающих 1-е и 5-е места
-- по количеству опубликованных вакансий
(
    SELECT
        e.name AS employer_name,
        COUNT(v.id) AS cnt
    FROM vacancies AS v
    JOIN employers AS e
        ON v.employer_id = e.id
    GROUP BY e.name
    ORDER BY cnt DESC
    LIMIT 1
)

UNION

(
    SELECT
        e.name AS employer_name,
        COUNT(v.id) AS cnt
    FROM vacancies AS v
    JOIN employers AS e
        ON v.employer_id = e.id
    GROUP BY e.name
    ORDER BY cnt DESC
    LIMIT 1 OFFSET 4
)
'''
```

```python
df = pd.read_sql_query(query_5_1, connection)
df
```

2. Напишите запрос, который для каждого региона выведет количество работодателей и вакансий в нём.
Среди регионов, в которых нет вакансий, найдите тот, в котором наибольшее количество работодателей.


```python
query_5_2 = f'''
-- Для каждого региона считаем количество работодателей и вакансий
-- Оставляем только те регионы, где вакансий нет,
-- и выбираем регион с максимальным числом работодателей
SELECT
    a.name AS region_name,
    COUNT(DISTINCT e.id) AS cnt_employers,
    COUNT(DISTINCT v.id) AS cnt_vacancies
FROM areas AS a
LEFT JOIN employers AS e
    ON a.id = e.area
LEFT JOIN vacancies AS v
    ON a.id = v.area_id
GROUP BY
    a.id,
    a.name
HAVING COUNT(DISTINCT v.id) = 0
ORDER BY cnt_employers DESC
LIMIT 1
'''
```

```python
df = pd.read_sql_query(query_5_2, connection)
df
```

```python
import requests

# Wiki link
url = 'https://ru.wikipedia.org/wiki/Города-миллионеры_России'

# Добавляем заголовок User-Agent, чтобы запрос выглядел как запрос из браузера
headers = {
    'User-Agent': (
        'Mozilla/5.0 (Windows NT 10.0; Win64; x64) '
        'AppleWebKit/537.36 (KHTML, like Gecko) '
        'Chrome/114.0.0.0 Safari/537.36'
    )
}

# HTML страничка
response = requests.get(url, headers=headers)

# Считываем таблицы со страницы и выбираем таблицу,
# содержащую столбец с названием города
tables = pd.read_html(response.text, match='Город')

# Нужная таблица первая по счету
df_cities = tables[0]

# Извлекаем названия городов-миллионников в список
million_cities = df_cities['Город'].tolist()

# Удаляем сноски из названий городов
million_cities = [city.split('[')[0] for city in million_cities]

print(million_cities)
```

```python
cities_tuple = tuple(million_cities)

query_5_7 = f'''
-- Подсчитываем количество вакансий компании Яндекс
-- по городам-миллионникам, а также добавляем общий итог
(
    SELECT
        a.name AS region_name,
        COUNT(v.id) AS cnt
    FROM vacancies AS v
    JOIN areas AS a
        ON v.area_id = a.id
    JOIN employers AS e
        ON v.employer_id = e.id
    WHERE e.name = 'Яндекс'
      AND a.name IN {cities_tuple}
    GROUP BY a.name
)

UNION ALL

(
    SELECT
        'Total' AS region_name,
        COUNT(*) AS cnt
    FROM vacancies AS v
    JOIN areas AS a
        ON v.area_id = a.id
    JOIN employers AS e
        ON v.employer_id = e.id
    WHERE e.name = 'Яндекс'
      AND a.name IN {cities_tuple}
)

ORDER BY cnt ASC;
'''
```

```python
df = pd.read_sql_query(query_5_7, connection)
df
```

При решении задачи примите во внимание следующее:
1. Рассматриваем только вакансии, у которых заполнено хотя бы одно из двух полей с зарплатой.
2. Если заполнены оба поля с зарплатой, то считаем зарплату по каждой вакансии как сумму двух полей, делённую на 2. Если заполнено только одно из полей, то его и считаем зарплатой по вакансии.
3. Если в расчётах участвует null, в результате он тоже даст null (посмотрите, что возвращает запрос select 1 + null). Чтобы избежать этой ситуацию, мы воспользуемся функцией [coalesce](https://postgrespro.ru/docs/postgresql/9.5/functions-conditional#functions-coalesce-nvl-ifnull), которая заменит null на значение, которое мы передадим. Например, посмотрите, что возвращает запрос `select 1 + coalesce(null, 0)`

Выясните, на какую зарплату в среднем может рассчитывать дата-сайентист с опытом работы от 3 до 6 лет. Результат округлите до целого числа. 

```python
query_6_6 = f'''
-- Рассчитываем среднюю зарплату для вакансий Data Scientist
-- с опытом работы от 3 до 6 лет
-- Если указаны обе границы зарплаты, берём их среднее значение;
-- если указана только одна, COALESCE подставляет её вместо пропущенной
SELECT 
    ROUND(AVG(
        (COALESCE(salary_from, salary_to) + 
         COALESCE(salary_to, salary_from)) / 2
    )) AS avg_salary
FROM vacancies
WHERE
    (
        name ILIKE '%data scientist%'
        OR name ILIKE '%data science%'
        OR name ILIKE '%исследователь данных%'
        OR (name LIKE '%ML%' AND name NOT LIKE '%HTML%')
        OR name ILIKE '%machine learning%'
        OR name ILIKE '%машинн%обучен%'
    )
    AND (salary_from IS NOT NULL OR salary_to IS NOT NULL)
    AND experience = 'От 3 до 6 лет';

'''
```

```python
# Фильтр вакансий Data Scientist по ключевым словам в названии.
# Исключаем HTML-вакансии при поиске по сокращению ML.
ds_filter = '''
(
    v.name ILIKE '%data scientist%'
    OR v.name ILIKE '%data science%'
    OR v.name ILIKE '%исследователь данных%'
    OR v.name ILIKE '%machine learning%'
    OR v.name ILIKE '%машинн%обучен%'
    OR (
        v.name ILIKE '%ml%'
        AND v.name NOT ILIKE '%html%'
    )
)
'''


# Выражение для расчета средней зарплаты:
# если указаны обе границы, берём среднее;
# если одна из границ отсутствует, COALESCE подставляет имеющееся значение.
salary_expr = '''
ROUND(AVG(
    (COALESCE(v.salary_from, v.salary_to) + COALESCE(v.salary_to, v.salary_from)) / 2.0
), 0)
'''


# Выражение для подсчета числа навыков в вакансии.
# Навыки в поле key_skills разделены табуляцией, поэтому
# количество навыков = количество разделителей + 1
skill_count_expr = '''
LENGTH(v.key_skills) - LENGTH(REPLACE(v.key_skills, CHR(9), '')) + 1
'''
```

```python
query_7_1 = f'''
-- Исследуем различия в средней зарплате Data Scientist
-- по регионам и требуемому опыту работы
-- Учитываем только вакансии с указанной зарплатой
-- и оставляем группы, где есть не менее 3 наблюдений

SELECT
    a.name AS area_name,
    v.experience,
    COUNT(v.id) AS vacancies_count,
    {salary_expr} AS avg_salary
FROM vacancies AS v
JOIN areas AS a
    ON v.area_id = a.id
WHERE
    {ds_filter}
    AND (v.salary_from IS NOT NULL OR v.salary_to IS NOT NULL)
GROUP BY
    a.name,
    v.experience
HAVING COUNT(v.id) >= 3
ORDER BY
    avg_salary DESC,
    vacancies_count DESC
'''
```

```python
df_7_1 = pd.read_sql_query(query_7_1, connection)
df_7_1
```

```python
# Выбираем регионы с наибольшим суммарным количеством DS-вакансий,
# чтобы визуализация была компактной и содержательной
top_areas_7_1 = (
    df_7_1.groupby('area_name', as_index=False)['vacancies_count']
    .sum()
    .sort_values('vacancies_count', ascending=False)
    .head(15)
)

top_area_names_7_1 = top_areas_7_1['area_name'].tolist()

df_7_1_top = df_7_1[df_7_1['area_name'].isin(top_area_names_7_1)]

# Строим сводную таблицу для тепловой карты:
# строки - регионы, столбцы - опыт, значения - средняя зарплата.
pivot_7_1 = df_7_1_top.pivot(
    index='area_name',
    columns='experience',
    values='avg_salary'
)

fig = px.imshow(
    pivot_7_1,
    text_auto=True,
    aspect='auto',
    title='Средняя зарплата Data Scientist по регионам и опыту'
)

fig.update_layout(
    xaxis_title='Опыт работы',
    yaxis_title='Регион'
)

fig.show()

fig.write_html('pivot_7_1.html')
```

```python
query_7_2 = f'''
-- Исследуем связь между количеством навыков,
-- требуемым опытом и средней зарплатой Data Scientist
-- Учитываем только вакансии с непустым списком навыков
-- и указанной зарплатой

SELECT
    v.experience,
    {skill_count_expr} AS skills_count,
    COUNT(v.id) AS vacancies_count,
    {salary_expr} AS avg_salary
FROM vacancies AS v
WHERE
    v.key_skills IS NOT NULL
    AND v.key_skills != ''
    AND {ds_filter}
    AND (v.salary_from IS NOT NULL OR v.salary_to IS NOT NULL)
GROUP BY
    v.experience,
    {skill_count_expr}
HAVING COUNT(v.id) >= 3
ORDER BY
    v.experience,
    skills_count
'''
```

```python
query_7_3 = f'''
-- Сравниваем востребованность и среднюю зарплату
-- для отдельных ключевых навыков в DS-вакансиях
-- Учитываем только навыки, встречающиеся не менее чем в 3 вакансиях

SELECT *
FROM (
    SELECT
        'Python' AS skill,
        COUNT(v.id) AS vacancies_count,
        {salary_expr} AS avg_salary
    FROM vacancies AS v
    WHERE
        v.key_skills ILIKE '%Python%'
        AND {ds_filter}
        AND (v.salary_from IS NOT NULL OR v.salary_to IS NOT NULL)

    UNION ALL

    SELECT
        'SQL' AS skill,
        COUNT(v.id) AS vacancies_count,
        {salary_expr} AS avg_salary
    FROM vacancies AS v
    WHERE
        v.key_skills ILIKE '%SQL%'
        AND {ds_filter}
        AND (v.salary_from IS NOT NULL OR v.salary_to IS NOT NULL)

    UNION ALL

    SELECT
        'PostgreSQL' AS skill,
        COUNT(v.id) AS vacancies_count,
        {salary_expr} AS avg_salary
    FROM vacancies AS v
    WHERE
        v.key_skills ILIKE '%PostgreSQL%'
        AND {ds_filter}
        AND (v.salary_from IS NOT NULL OR v.salary_to IS NOT NULL)

    UNION ALL

    SELECT
        'Machine Learning' AS skill,
        COUNT(v.id) AS vacancies_count,
        {salary_expr} AS avg_salary
    FROM vacancies AS v
    WHERE
        v.key_skills ILIKE '%Machine Learning%'
        AND {ds_filter}
        AND (v.salary_from IS NOT NULL OR v.salary_to IS NOT NULL)

    UNION ALL

    SELECT
        'Pandas' AS skill,
        COUNT(v.id) AS vacancies_count,
        {salary_expr} AS avg_salary
    FROM vacancies AS v
    WHERE
        v.key_skills ILIKE '%Pandas%'
        AND {ds_filter}
        AND (v.salary_from IS NOT NULL OR v.salary_to IS NOT NULL)
) AS skills_data

WHERE vacancies_count >= 3
ORDER BY avg_salary DESC
'''
```

При этом более высокие уровни средней зарплаты наблюдаются для комбинаций навыков, связанных с инфраструктурой хранения данных, например SQL и PostgreSQL. Это может свидетельствовать о повышенной ценности специалистов, работающих на стыке анализа данных и инженерии данных.

```python
query_7_5 = f'''
-- Определяем индустрии, в которых чаще всего встречаются
-- вакансии для Data Scientist

SELECT
    i.name AS industry_name,
    COUNT(DISTINCT v.id) AS vacancies_count
FROM vacancies AS v
JOIN employers AS e
    ON v.employer_id = e.id
JOIN employers_industries AS ei
    ON e.id = ei.employer_id
JOIN industries AS i
    ON ei.industry_id = i.id
WHERE
    {ds_filter}
GROUP BY
    i.name
ORDER BY
    vacancies_count DESC
LIMIT 15
'''
```

```python
df_7_5 = pd.read_sql_query(query_7_5, connection)
df_7_5
```

```python
df_7_5_top = df_7_5.head(10).copy()

# Длинные названия индустрий упрощаем для удобства визуализации
df_7_5_top['industry_short'] = [
    'Разработка ПО',
    'Автоматизация / IT-консалтинг',
    'Банк',
    'Медиа / интернет-ресурсы',
    'Мобильная связь',
    'Маркетинг / реклама / PR',
    'Управляющая компания',
    'Консалтинг',
    'Фиксированная связь',
    'Наука / исследования'
]

fig = px.bar(
    df_7_5_top.sort_values('vacancies_count', ascending=True),
    x='vacancies_count',
    y='industry_short',
    orientation='h',
    text='vacancies_count',
    title='Топ-10 индустрий с наибольшим спросом на Data Scientist'
)

fig.update_layout(
    height=550,
    xaxis_title='Количество вакансий',
    yaxis_title='Индустрия'
)

fig.show()

df_7_5_top[['industry_short', 'industry_name', 'vacancies_count']]

fig.write_html('plots/pivot_7_5.html')
```
